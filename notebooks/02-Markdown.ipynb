# Introducción a Markdown

Las celdas de markdown nos permiten incluir textos de forma sencilla, usando determinadas marcas para indicar niveles de encabezado, estilos tipográficos, etc... que se reflejarán en la salida (En este caso, a HTML para que lo muestrer el nevagador, pero tambien a PDF, Latex, etc...)

La filosofía es que sea tan fácil de escribir como de leer. Pero sobre todo se prima la legibilidad. Un texto en markdown deberia ser perfectamente legible aun como texto simple, sin ningun formateo. Las marcas usadas son muy sencillas y se mapean directamente con un subconjunto limitado de Html.

Podemos incluir texto HTML directamente dentro del documento markdown, como <code><i>aquí</i></code>, aunque no es lo recomendado.

## Encabezados

Los encabezados se pueden obtener usando varios caracteres # al principio de la línea, tantos como sea el nivel del encabezado:

A modo de ejemplo, el resultado de escribir esto:

    # Cabecera de nivel 1: Título principal
    ## Cabecera de nivel 2: Sección
    ### Cabecera de nivel 3: Subsección
    #### Cabecera de nivel 4: Y así ...
    ##### Cabecera de nivel 5: ... hasta 6 ...
    ###### Cabecera de nivel 6: ... subniveles
    
    
Seria:

# Cabecera de nivel 1: Título principal
## Cabecera de nivel 2: Sección
### Cabecera de nivel 3: Subsección
#### Cabecera de nivel 4: Y así ...
##### Cabecera de nivel 5: ... hasta 6 ...
###### Cabecera de nivel 6: ... subniveles

Tambien se acepta una sintaxis alternativa, consistente en usar caracteres `=` (para cabeceras de primer nivel) o `-` para cabeceras de segundo nivel, el siguiente texto literal:

    Cabecera principal
    ==================
    
    Cabecera Secundaria
    -------------------
    
Se muestra así:

Cabecera principal
==================
    
Cabecera Secundaria
-------------------

## Listas

Markdown soporta listas con números o sin números (con puntos). Para las listas no ordenadas, empezamos cada línea de la lista con uno de los siguientes caracteres: `*`, `+` o  `-`:

    * Rojo
    * Verde
    * Azul
    
Se vería

* Rojo
* Verde
* Azul

Para las listas numeradas, simplemente usar numeros seguidos por un punto. Pero **Ojo, porque en la salida Html no se tendrán en cuenta el valor de esos números, se cambiaran por elementos dentro de una lista que se autonumera sola**. Es decir que la siguiente lista:

    1. Linux
    2. Mac
    3. Windows
    
Producen la misma salida que:

    1. Linux
    1. Mac
    1. Windows
    
E incluso que:

    2. Linux
    9. Mac
    37. Windows
    
Es decir:

2. Linux
9. Mac
37. Windows

Esto quiza cambie en el futuro, así que sugiero numerar las listas de forma secuencial, 1, 2, 3, etc... para evitarnos posibles sorpresas.

Atención porque es posible que por causa de esto se interprete como una lista un texto 
que no debería serlo, por ejemplo:

>    1966\. ¡Qué buena cosecha!

Que se interpretaría:

1966. ¡Qué buena cosecha!

Para evitar esto, se puede escapar el punto:

    1966\. ¡Qué buena cosecha!


Los marcadores de listas deben empezar cada línea, pero pueden ser indentados hasta un máximo de tres espacios, y se pueden incluir listas dentro de listas:

     * Seccion 1
       - Seccion 1.1

## Incluir código

Podemos usar una indentacion de al menos 4 espacios para incluir código fuente u otros textos que no deseamos que sean interpretados por markdown, sini que sean interpretados literalmente.

    for i in range(10):
        print('hola')

Un bloque de código continua hasta que se alcanza una línea de texto no indentada (o lleguemos al fin del contenido).

Dentro del bloque de código, los símbolos _ampersand_ (`&`) y los símbolos de mayor/menor (`<` y `>`) se convierten de forma automática a entidades Html, lo que hace que sea muy cómodo
incluir un fragmento de Html, simplemente hay que pegarlo e indentarlo. Por ejemplo:

    <div class="footer">
        &copy; 2004 Foo Corporation
    </div>

La sintaxis general de Markdown no funciona dentro de los bloques de código, es decir
que los asteriscos solo son asteriscos literales. Esto hace muy fácil describir el formato
de Markdown en Markdown, como se hace en este _notebook_ :-).

## Enlaces

El formato Mardown  soporta dos tiops de enlaces: En línea o referencias. En ambos casos, el texto del enlace debe ir delimitado por corchetes (`[` y `]`)

### Enlaces en línea

Para crear un enlace en línea, hay que añadir, imdiatemante despues del corchete de cierre del texto del enlace, la URL destino entre paréntesis. Con un ejemplo se entenderá mejor:

    La [página principal de Python](http://www.python.org/) es muy interesante
    
Se verá

> La [página principal de Python](http://www.python.org/) es muy interesante
   
Podemos añadir dentro del paréntesis del enlace, separado por un espacio y rodeados por comillas, el título del enlace. El siguiente ejemplo:

    La [página principal de Python](http://www.python.org/ "Python homepage").

Produce:

La [página principal de Python](http://www.python.org/ "Python homepage").

(Se puede mostrar el título de enlace poniendo el puntero sobre el mismo y, sin pulsarlo, esperar unos segundos)

### Enlaces por referencia

En los enlaces por referencia, añadimos un segundo bloque de corchetes, dentro del cual ponemos una referencia que usaremos para identificar el enlace, es decir:

    [texto del enlace][ref]
    
Por ejemplo:

    La [página principal de Python][python] es muy interesante
    
Que se vería:

La [página principal de Python][python] es muy interesante
    
Y luego, en alguna parte del documento, se define el enlace vinculado con la etiqueta, usando el siguiente formato:

    [ref]: url "título opcional"
    
Por ejemplo:

    [python]: http://www.python.org/ "La página principal del lenguaje Python"

Esa línea es invisible, asi que podemos ponerla donde queramos, lo normal es o ponerlos cerca de donde se usan o reservar una parte del documento, por ejemplo al final, para poner allí todos los enlaces.

[python]: http://www.python.org/ "La página principal del lenguaje Python"

Otra forma de incluir enlaces es incluir una url entre simbólos de mayor/menor. este realizará el enlace de forma inmediata:

    La página web de Python: <http://www.python.org/>.
    
Que se vería:

La página web de Python: <http://www.python.org/>.

Los enlaces por referencia son un poco más difíciles de escribir, pero hacen que el documento sea mucho más fáciles de leer.

[python]: http://www.python.org/ "La página principal del lenguaje Python"

## Énfasis al texto: negritas, itálicas, etc...

Los caracteres `*` y  `_` sirven de indicadores de énfasis en Markdwon. Los textos rodeados por un caracter `*` o `_` se resaltan con itálicas (usando el tag HTML `em`). Los textos rodeados por dobles caracteres `**` o `__` se resaltan en negrita (Con el tag HTML `strong`). Es decir, el siguiente texto:

    *un solo asterisco*

    _un solo subrayado_

    **dobles asteriscos**

    __dobles subrayados__

Se mostraría:

*un solo asterisco*

_un solo subrayado_

**dobles asteriscos**

__dobles subrayados__

Si queremos usar un asterisco o caracter subrayado literalmente, que Markdown interpreta como un  indicar de énfasis, podemos escaparlo con una barra invertida:

    Este texto \*entre asteriscos no se interpreta como
    énfasis\* porque escapamos el significado
    
Que se vería:

Este texto \*entre asteriscos no se interpreta como 
énfasis\* porque escapamos el significado

Podemos representar un fragmento de código (Normalmente se representa con letra monoespaciada) usando el caracter _comillas invertidas_ (\`), como en el siguiente ejemplo:

    La funcion `print()` muestra sus parámetros en pantalla
    
Que se muestra:

La funcion `print()` muestra sus parámetros en pantalla

Se pueden usar una o dos comillas invertidas:

    La funcion ``print()`` muestra sus parámetros en pantalla
    
Que se muestra:

La funcion ``print()`` muestra sus parámetros en pantalla



## Imágenes

La sintaxis para incluir imágenes es similar a la de los enlaces, permitiendo dos estilos, en línea o por referencia.

### Imágenes en línea

La sintaxis es:

    ![Alt text](/path/to/img.jpg)

    ![Alt text](/path/to/img.jpg "Optional title")

es decir, igual que los enlaces pero incluyendo la marca `!`. El siguiente texto:

    ![Einstein](./img/einstein-sticker.jpg "Science rules!")

Mostrará:

![Einstein](./img/einstein-sticker.jpg "Science rules!")

El texto entre corchetes se usará como alternativa si no se puede mostrar la imagen (por ejempllo, por que no la encuentra). Si incluimos un texto opcional entre comillas después del url, se usará como atributo `title` de la misma:

### Imágenes por referencia

Similar a los enlaces, incluimod dos bloques con corchetes, el segundo sera una referencia, y luego vinculamos la referencia con la url de la imagen con una  línea aparte:

    ![¡Atención'][warning]
    
    [warning]: ./img/warning-icon.png

Se ve:

![¡Atención'][warning]
    
[warning]: ./img/warning-icon.png

Se puede incluir código, incluyendo coloreado de sintaxis, usando
tres comillas invertidas antes del bloque. Si justo despues de las comillas ponemos el

    ```python

    ```

```python
# Ejemplo en Python
print("Hola, Mundo")
```
