```python
import os
```

```python
tipos = ('.pdf', '.ipynb', '.svg', '.png', '.R')
data = {}
for (dir_base, dirs, files) in os.walk('..'):
    for fn in files:
        (_, ext) = os.path.splitext(fn)
        if ext in tipos:
            full_name = os.path.join(dir_base, fn)
            size = os.path.getsize(full_name)
            if ext in data:
                data[ext] = data[ext] + size
            else:
                data[ext] = size
data
        
```

```python
%matplotlib notebook
import pandas as pd

s = pd.Series(data)
s.plot.barh()
```
