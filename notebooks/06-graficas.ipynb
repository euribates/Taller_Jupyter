Gráficas con matplotlib y seaborn
==========================================

```python
import math
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

Gráficas con plot
-----------------------------

Plot dibuja líneas o marcadores en un sistema de ejes. plot acepta un número variable de argumentos, que representara todos sobre la misma grafica:

```python
x = np.linspace(-math.pi, math.pi, 1000)  # definimos un espacio lineal
y = np.sin(x)
plt.plot(x, y)
pass  # esta línea evita que se imprima el resultado
```

Podemos representar más de una serie de datos:

```python
x = np.linspace(-2*math.pi, 2*math.pi, 1000)  # definimos un espacio lineal
y = np.sin(x)
y2 = np.sin(x ** 2 )
plt.plot(x, y, x, y2)
pass
```

Podemos representar los ejes x o y en el punto 0 (O en el que queramos) con los métodos `axhline` y `axvline`. Indicamos la posición en la que queremos que se pinte el eje. opcionalmente podemos indicar el color:

```python
x = np.linspace(-2*math.pi, 2*math.pi, 1000)  # definimos un espacio lineal
y = np.sin(x) - 0.25
y2 = np.sin(x ** 2 ) + 0.25
plt.plot(x, y, x, y2)
plt.axhline(0, color='#996633')
plt.axvline(0, color='#996633')
pass
```

```python
El tipo de línea, marcas y colores se pueden determinar con una seríe de cadenas de formato
```

```python
plt.plot([1,2,3], 'go-', label='line 1', linewidth=1)
plt.plot([1,4,9], 'rs',  label='line 2')
plt.axis([0, 4, 0, 10])
plt.legend();
```

```python
### Histogramas
```

Podemos usar la funcion `hist` de la librería para crear un gráfico de histograma.

```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 10, 5
x = mu + sigma * np.random.randn(10000)

# Histograma con 50 barras
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)

plt.xlabel('Error medida')
plt.ylabel('Probabilidad')
plt.title('Histograma')
plt.text(-13, .084, r'$\mu=100,\ \sigma=15$')
plt.axis([-15, 30, 0, .1])
plt.grid(True)
plt.show()
```

Uso de Seaborn
----------------------

Podemos usar la libreria seaborn para que las gráficas sean más atractivas. Por ejemplo, si partimos de esta función:

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

Por defecto matplotlib general esto:

```python
sinplot()
```

Si nos pasamos a la configuración por defecto de seaborn, el estilo de la gráfica cambia. Para ello usamos una llamada a `set()`:

```python
sns.set()
sinplot()
```

HAy cinco temas predefinidos en seaborn ahora mismo: `darkgrid`, `whitegrid`, `dark`, `white` y `ticks`. Cada uno de ellos está ajustado para diferentes aplicaciones y preferencias personales. El tema por defeceto es `darkgrid`. La parrilla se incluye para ayudar con la referencias de los ejes, pero se mantienen en un tono gris claro y con lineas blancas para que no compitan en interes con las líneas que representan datos. El estilo `whitegrid` es similar, pero está diseñado para funcionar mejor con gráficas con muchos datos:



```python
sns.set_style('whitegrid')
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```
