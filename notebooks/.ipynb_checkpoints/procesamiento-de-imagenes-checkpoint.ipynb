# Procesamiento de imagenes

## Con Pillow, Tesseract

Primero de todo, comprobemos que tenemos instaladas las librerias de [Pillow](https://python-pillow.org/). EN caso de no tenerlas instaladas, es tan sencillo como:

    pip install pillow

```python
import PIL
```

**Pillow** es un _fork_ de una librería anterior que se llamaba PIL (Python Image LIbrary). Se sigue importando con
el nombre de PIL por razones de compatibilidad.

Vamos a cargar una imagen, para ello usaremos la clase `Image` y el método de clase `open`. 

```python
from PIL import Image

im = Image.open('pruebas.png')
im
```

```python
from PIL import ImageFilter

ib = im.convert('RGBA')
ib.filter(ImageFilter.BLUR).rotate(90)
```

```python
i2 = im.crop((0,318, 1024, 410))
i2
```

```python
ir = i2.rotate(0.89)
ir
```

Veamos si está importado `pytesseract`. Es un _rapper_ sobre una herramienta de OCR llamada Tesseract, que tambien tendría que estar instalada.

```python
import pytesseract
```

```python
i2
```

```python
s = pytesseract.image_to_string(i2, lang='spa')
print(s)
```

```python
# Si le añadimos la ayuda de decirle qué lenguaje es, mejoran los resultados

s = pytesseract.image_to_string(i2, lang='deu')
print(s)
```
