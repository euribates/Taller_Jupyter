## Cálculo aproximado de π 

Usando el método Montecarlo. Vamos a generara aleatoriamente puntos en un recuadro que vaya desde 0 a 1, tanto en el eje $x$ como en el eje $y$.

Sabemos al area de un cuadrado, dado su lado $r$:

$$ A_1 = r^2 $$

Y también sabemos que el área del círculo de radio $r$:

$$ A_2 = \pi r^2 $$

Como estamos trabajando solo en el cuadrante superior,  el area sería un cuarto de la del círculo completo, es decir:

$$ A_3 = \frac{\pi r^2}{4} $$

Por lo tanto, la relación entre el área del cuadrado y el área del cuarto de círculo es:

$$ \frac{A_3}{A_1} = \frac{\pi r^2}{4r^2} = \frac{\pi}{4}$$

Como el número de puntos totales es proporcional al área del cuadrado, y el número de puntos que caen dentro del círculo es proporcional al área del cuarto de círculo, podemos obtener la estimación de $\pi$ haciendo:

$$ \pi \approx 4\cdot\frac{[puntos\: dentro\: del\: círculo]}{[puntos\: totales]} $$


```python
# Para que las gráficas esten embebidas
%matplotlib inline
```

```python
import math
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import random
```

```python
N_DISPAROS = 5000


x = np.array([random.random() for n in range(N_DISPAROS)])
y = np.array([random.random() for n in range(N_DISPAROS)])

plt.plot(x, y, 'b.');
```

```python
inside = np.sqrt(x**2 + y**2) < 1
plt.plot(x[inside], y[inside], 'r.');
```

```python
outside = np.invert(inside)
plt.plot(x[outside], y[outside], 'b.');
```

```python
plt.plot(x[inside], y[inside], 'rx')
plt.plot(x[outside], y[outside], 'bo')

space = np.linspace(0, math.pi/2, 100)
plt.plot(np.cos(space), np.sin(space), 'w-');
```

```python
approx_pi = 4*sum(inside) / N_DISPAROS
print('El valor aproximado de pi con {} intentos es: {}'.format(
    N_DISPAROS,
    approx_pi,
    ))
print('El error es: {}'.format(math.pi - approx_pi))
```
